Интерактивное приложение Dash, где можно выбирать столбец из выпадающего списка, и график будет обновляться соответственно.

```python
# Создание экземпляра Dash
app = dash.Dash(__name__)

# Определение макета приложения
app.layout = html.Div([
    html.H1('Interactive Graph'),
    dcc.Graph(id='interactive-plot'),
    html.Label('Select a column to plot:'),
    dcc.Dropdown(
        id='dropdown-column',
        options=[{'label': col, 'value': col} for col in df.columns],
        value='engine-size'
    )
])

# Определение обработчика для обновления графика при выборе нового столбца
@app.callback(
    Output('interactive-plot', 'figure'),
    [Input('dropdown-column', 'value')]
)
def update_plot(selected_column):
    fig = px.scatter(df, x=selected_column, y='price', title=f'{selected_column.capitalize()} vs Price')
    return fig

# Запуск приложения
if __name__ == '__main__':
    app.run_server(debug=True)
```
